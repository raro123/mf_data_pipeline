# Automated Authentication Flow for Zerodha Kite Connect

Zerodha's Kite Connect API requires a manual login to obtain the request token due to security reasons. This means full automation of the authentication process is not possible unless Zerodha provides a refresh token or headless login API. However, you can streamline the workflow by:

- Storing API credentials in a `.env` file
- Saving the access token after the first manual login for reuse (until it expires)
- Prompting for the request token only when needed

The following cell demonstrates a semi-automated authentication flow.

```python
import os
from dotenv import load_dotenv
from kiteconnect import KiteConnect

load_dotenv()
API_KEY = os.getenv('API_KEY')
API_SECRET = os.getenv('API_SECRET')
ACCESS_TOKEN = os.getenv('ACCESS_TOKEN')

kite = KiteConnect(api_key=API_KEY)

if ACCESS_TOKEN:
    kite.set_access_token(ACCESS_TOKEN)
    print('Access token loaded from .env')
else:
    print('Visit this URL and login to get your request token:')
    print(kite.login_url())
    REQUEST_TOKEN = input('Enter the request token: ')
    data = kite.generate_session(REQUEST_TOKEN, api_secret=API_SECRET)
    ACCESS_TOKEN = data['access_token']
    # Optionally, save the access token for future use
    with open('.env', 'a') as f:
        f.write(f'\nACCESS_TOKEN={ACCESS_TOKEN}\n')
    kite.set_access_token(ACCESS_TOKEN)
    print('Access token set and saved to .env')
```

```python
import os

def download_all_mf_instruments(kite: KiteConnect, save_path: str = None):
    """
    Download the list of all mutual fund instruments using Zerodha Kite API.
    Optionally save the result as a CSV file.
    Returns a pandas DataFrame.
    """
    try:
        mf_instruments = kite.mf_instruments()
        df = pd.DataFrame(mf_instruments)
        if save_path:
            dir_name = os.path.dirname(save_path)
            if dir_name and not os.path.exists(dir_name):
                os.makedirs(dir_name, exist_ok=True)
            df.to_csv(save_path, index=False)
            print(f"Mutual fund instruments saved to {save_path}")
        return df
    except Exception as e:
        print(f"Error downloading mutual fund instruments: {e}")
        return None
```

```python
download_all_mf_instruments(kite, save_path='data/mf/mf_instruments.csv')
```
